# **1. Money Change**

```python
# Uses python3
def get_change(n):
  count = 0
  for i in [10, 5, 1]:
      if n>=i:
          q = n//i
          count += q
          n = n%i
          if n==0:
            n=count        
            return n

n = int(input())     
print(get_change(n))
```

# **2. Maximum Value of the Loot**

```python
# Uses python3
def get_optimal_value(capacity, lst):
    value = 0.
    # write your code here
    for v,w in lst:
      if capacity==0:
          return value
      else:
        a = min(w, capacity)
        value += a*v/w
        capacity -= a
    return value

n, bigW = map(int, input().split())
lst = []
for _ in range(n):
    v, w = map(int, input().split())
    if v == 0:
        continue
    lst.append((v, w))

lst.sort(key = lambda x: x[0]/x[1], reverse = True)

opt_value = get_optimal_value(bigW, lst)
print("{:.10f}".format(opt_value))
```

# **3. Car Refueling**

```python
# Uses python3
def compute_min_refills(distance, tank, stops):
    numRefills = 0
    currentPosition = 0

    while(currentPosition <= stops):
        lastPosition = currentPosition
        while(currentPosition <= stops  and  distance[currentPosition + 1] - distance[lastPosition] <= tank):
          currentPosition = currentPosition+1

        if (currentPosition == lastPosition): return -1 
        if (currentPosition <= stops): numRefills = numRefills+1
    return numRefills

distance = []
pointA = 0
distance.append(pointA)
pointB = int(input())
tank = int(input())
stops = int(input())
templist = list(map(int, input().split()))
for x in templist:
  distance.append(x)

distance.append(pointB)

print(compute_min_refills(distance, tank, stops))
```

# **4. Maximum Advertisement Revenue**

```python
#Uses python3
def max_dot_product(a, b, n):
    #write your code here
    res = 0
    for i in range(n):
        res += a[i] * b[i]
    return res

n = int(input())
a = list(map(int, input().split()))
b = list(map(int, input().split()))
a.sort()
b.sort()
print(max_dot_product(a, b, n))
```

# **5. Collecting Signature**

```python
#Uses python3
def optimal_points(lst, n):
    index = 0
    coordinates = []

    while index < n:
        curr = lst[index]
        while index < n-1 and curr[1]>=lst[index+1][0]:
            index += 1
        coordinates.append(curr[1])
        index += 1
    return coordinates

n = int(input())
lst = []

for _ in range(n):
    a, b = map(int, input().split())
    lst.append((a,b))

lst.sort(key = lambda x: x[1])

points = optimal_points(lst, n)
print(len(points))
print(*points)
```
